"Prediction of the presence or absence of heart disease in Cleveland based on the age, serum cholestrol and maximum heart rate of the patient"

According to clinical trials and research, Coronary heart disease is the main cause of death in the western world. Therefore, it is essential to identify its risk factors in early stages of life so that proper treatments can be recruited.
A research about heart diseases and its risk factors was caried out on the population of cleveland. This project will explore these  risk factors and if they can predict the presence of heart disease. 

```r
#First, the necessary libraries have to be loaded so the coding in R can properly take place.
library(tidyverse)
library(repr)
library(readxl)
library(tidymodels)
library(readr)
library(dplyr)
options(repr.matrix.max.rows=303)
```

```r
#The data set has to be loaded into R.
#The dataset does not have any column names and therefore, it can be read using delim with col_names= FALSE.

cleve_data<-read_delim("data/heart_disease/processed.cleveland.data", delim= ",",
                      col_names = FALSE)
cleve_data
```

In the next step, ithe columns are identified, using the websit in which the data was imported from in the first place.

```r
cleve_data <- rename(cleve_data,
                     age= X1,
                     sex= X2,
                     cp= X3,
                     trestbps= X4,
                     chol= X5,
                     fbs= X6,
                     restecg= X7,
                     thalach= X8,
                     exang= X9,
                     oldpeak= X10,
                     slope= X11,
                     ca= X12,
                     thal= X13,
                     num= X14)

cleve_data
                     
```

According to the dataset, there are 14  columns (variables) and 303 rows (obervations). The "num" column in the "cleve_data" dataset indicates the presense (1,2,3,4) or the absence (0) of heart disease in that specific observation.

The predictive question being asked in this project is as follow:
" Will the age, serum cholestrol (chol) and the maximum heart rate(thalach) of the patient determine if they will be dealing with heart disease?"

```r
cleve_select <- cleve_data |>
    select(age, chol, thalach, num)
cleve_select
```

The above table is the table we will be working with in this project.

```r
cleve_tidy <- cleve_select |>
    mutate(num = as_factor(num)) |> #we have to make sure that the classification column is a factor.
    mutate(num = fct_recode(num, 
                            "Absent" = "0", 
                            "Present" = "1", 
                            "Present" = "2",
                            "Present" = "3", 
                            "Present" = "4" )) |>
#according to the website, where num is zero, no heart disease exist and where num is 1-4, the patient has heart disease.
    rename(heart_disease = num)
    
    
cleve_tidy
```

Using classification and splitting the data into a training set and a testing set, the presence or absence of the heart disease can be predicted in any given observation.

```r
#Here, we first set the seed equal to 1, just as a random number.
set.seed(1)
#Then we have to split the data, having in mind that there are 4 possibilities for the presence of the heart disease and 1 possiblity for its absence,
#hence, a ration of 4:1 is resulted.
cleve_split <- initial_split(cleve_tidy, prop=0.8, strata = heart_disease)
#Then, the dataset has to be split into the training and the testing set.
cleve_train <- training(cleve_tidy)
cleve_test <- testing(cleve_tidy)

#In the next step, the neighbors are set.
knn_spec <- nearest_neighbor (weight_func= "rectangular", neighbors = 5) |>
    set_engine ("kknn") |>
    set_mode ("classification")

cleve_recipe <- recipe (heart_disease ~ age + chol + thalach, data= cleve_tidy) |>
    step_center (all_predictors())
    step_scale (all_predictors ())

cleve_fit <- workflow () |>
    add_model (knn_spec) |>
    add_recipe (cleve_recipe) |>
    fit (data=cleve_train)

#In this last step, "..." represents a new observation for which the presence or absence of the disease will be prediceted.
#cleve_predict <- cleve_fit |>
#    predict ( ... )
#cleve_predict
```
